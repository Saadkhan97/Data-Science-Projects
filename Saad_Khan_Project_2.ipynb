```python
# Extracting categorical columns:
catFeatures= [col for col in data.columns if col in data.select_dtypes(include=object).columns]

# # Extracting All Features:
features = [col for col in data.columns if col not in ['is_claim']]

# # Printing Count of Features:
# print(f'total features: {len(features)}')
# # Printing Name of  Features:
print(features)
```

```python
catFeatures
```

```python
# Spliting Features and Target Variable:
X, y = data.loc[:, features], data.loc[:, 'is_claim']
# Checking Dataset Shape (Features):
print(X.shape)
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder

# Encoding Categorical Data
labelEncode = LabelEncoder()

# Iterating Over each categorial features:
for col in catFeatures:
    # storing its numerical value:
    X[col] = labelEncode.fit_transform(data[col])
```

```python
X
```

```python
from sklearn.preprocessing import OneHotEncoder

onehot = OneHotEncoder()

transformed = onehot.fit_transform(data[["fuel_type"]])
print(transformed.toarray())
```

```python
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), ["fuel_type"])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```
